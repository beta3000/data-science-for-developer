```python
x = dataset.drop(['PassengerId', 'Cabin', 'Ticket', 'Fare', 'Parch', 'SibSp'], axis=1)
y = x.Survived
x = x.drop(['Survived'], axis=1)
x.head(15)
```

```python
from sklearn.preprocessing import LabelEncoder
labelEncoder_x = LabelEncoder()
x.Sex = labelEncoder_x.fit_transform(x.Sex)
```

```python
x.head()
```

```python
sum(x.Embarked.isnull())
```

```python
row_index = x.Embarked.isnull()
x.loc[row_index, 'Embarked'] = 'S'
```

```python
x.Embarked = labelEncoder_x.fit_transform(x.Embarked)
```

```python
x.head()
```

```python
x = x.drop(['Name', 'Age'], axis = 1)
x.head()
```

```python
#Random forest
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 100, criterion = 'entropy', random_state=0)
```

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator=classifier, X=x, y=y, cv=10)
print("Random forest accuracy:", accuracies.mean(), accuracies.std())
```

```python
from sklearn.linear_model import LogisticRegression
classifier2 =  LogisticRegression(penalty = '12', random_state=0)
accuracies = cross_val_score(estimator=classifier2, X=x, y=y, cv=10)
print("LR accuracy:", accuracies.mean(), accuracies.std())
```
